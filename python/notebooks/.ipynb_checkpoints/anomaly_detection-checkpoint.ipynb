# anomaly detection — summary

## method
one-class support vector machine trained on queenright data only (all hives, all hours).
the model learns the "normal" feature envelope; queenless observations are scored as novelty.

per-time-slice scoring: one scaler + one detector fitted per hour slice,
so the model does not confuse day/night variation with anomaly.

the feature are filtered to keep only the ones whose zscore diverge between queenright and queenless data

## evaluation
- anomaly score histograms: visual check that queenless scores shift left (more anomalous)
- mann–whitney u test + area under the receiver operating characteristic curve: 
  non-parametric measure of how separable the two populations are
- cohen's d: effect size in pooled standard deviations

## mosaic visualisation
rows = hourly observations, columns = features (z-scored against queenright baseline).
left strip = anomaly score (darker = more anomalous, red border = bottom 1st percentile).
baseline panel shows the same time window aggregated across queenright hives
(mean, worst, furthest from centroid, or single hive selectable).
columns are reordered by feature similarity for visual clustering.

## findings
- hive 04 queenless period is well separated from baseline
- hive 03 queenless period shows weaker separation if any
- mosaic heatmap highlight which features deviate most during queenlessness
  - anomaly score is displayed vertically and shows at what hour the data diverge from baseline
  - if dsipalyed via matplotlib :
    - columns are cluster according to common information
    - feature from the same group share same color scale

```python
import dataframes
import anomalies
import paths
import numpy
import pandas
```

```python
all_features = dataframes.from_filepath(paths.all_merged_features_filepath)
all_features = all_features.drop_duplicates(subset=["timestamp", "time_slice", "hive"], keep="first")
all_features =all_features.dropna()
```

## focused periods of the day
I make the educated guess that activity is higher during daylight

```python
selected_time_slices = [
    # '10-11', 
    '11-12', 
    '12-13', 
    '13-14', 
    # '14-15', 
    # '15-16', 
    # '16-17', 
    # '17-18',
]
```

```python
all_features =all_features[
    all_features['time_slice'].isin(selected_time_slices)
]
```

## split data into queenless and queenright
- we split queenless data per hive
- queenrght data is kept as is

```python
queenright_features_dataframe = all_features[all_features['queenlessness'] == False]
queenless_features_dataframe = all_features[all_features['queenlessness'] == True]
```

```python
queenless_features_dataframe_hive_03 = queenless_features_dataframe[
    queenless_features_dataframe['hive'] == "hive_03"
]
queenless_features_dataframe_hive_04 = queenless_features_dataframe[
    queenless_features_dataframe['hive'] == "hive_04"
]
```

## features selection
keeping only features/columns which shows divergence between queenless and queenright data

```python
def most_anomalous_feature_columns(outlier_dataframes, inlier_dataframe):
    inlier_zscores = dataframes.zscored_dataframe_from_dataframe_and_baseline(
        inlier_dataframe, inlier_dataframe,
    )
    inlier_means = inlier_zscores.mean()

    anomalous_columns = set(inlier_zscores.columns)
    for outlier_dataframe in outlier_dataframes:
        outlier_zscores = dataframes.zscored_dataframe_from_dataframe_and_baseline(
            outlier_dataframe, inlier_dataframe,
        )
        sign_product = outlier_zscores.mean() * inlier_means
        divergent_columns = set(sign_product[sign_product < 0].index)
        anomalous_columns = anomalous_columns.intersection(divergent_columns)

    return list(anomalous_columns) + ['timestamp', 'hive', 'time_slice','queenlessness', 'temperature',
       'humidity']
```

```python
preselected_feature_columns = most_anomalous_feature_columns(
    [
        queenless_features_dataframe_hive_03,
        queenless_features_dataframe_hive_04,
    ],
    queenright_features_dataframe
)
```

```python
preselected_feature_columns
```

```python
queenright_features_dataframe = queenright_features_dataframe[preselected_feature_columns]
queenless_features_dataframe = queenless_features_dataframe[preselected_feature_columns]
queenless_features_dataframe_hive_03 = queenless_features_dataframe_hive_03[preselected_feature_columns]
queenless_features_dataframe_hive_04 = queenless_features_dataframe_hive_04[preselected_feature_columns]
```

## anomaly detector choice
*output do not really vary between different detectors*

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM

anomaly_detectors = {
    'isolation_forest': IsolationForest(random_state=42),
    'local_outlier_factor': LocalOutlierFactor(novelty=True),
    'svm': OneClassSVM(),
}
```

```python
def anomaly_scores_from_dataframe_from_inliers_dataframe(inliers_dataframe):
    scorers = anomalies.scorers_from_inliers_dataframe_and_detector(
        inliers_dataframe,
        anomaly_detectors['isolation_forest'],
    )
    return lambda dataframe: anomalies.anomaly_scores_from_dataframe_and_scorers(dataframe, scorers)
```

```python
anomaly_scores_from_dataframe = anomaly_scores_from_dataframe_from_inliers_dataframe(
    queenright_features_dataframe
)
```

## anomaly discrimination test evaluation
we provide 2 dataframes, baseline and investigated 
### left graph
we compare the anomaly score of both data as distributions
negative values (toward left) are more abnormal/outlier  
thus the more the distribution is distributed to the left the more abnormal  
the vertical line, median is a good single indicator of it
### right graph
Evaluate hwo reliable it is to assert if a value is normal or not, based on: 
- the two data distributions
- the tested value anomaly score

```python
anomalies.discrimination_figure_from_investigated_scores_and_baseline_scores(
    anomaly_scores_from_dataframe(queenless_features_dataframe),
    anomaly_scores_from_dataframe(queenright_features_dataframe),
);
```

## mosaic monitoring
we show each selected feature anomaly score for each hour for both queenless and queenright data  
- anomaly score on the left (positive = normal , negative = abnormal)
- features colored by type
- features clustered by redundance
since all queenrigt data are expected to be given as baseline a aprameter indicate for each hour which score is shown:
- worst -> the lowest anomaly score
- best -> the highest anomaly score
- furthest -> the fursthest from the investigated queenless data at that hour
- a number -> that beehive
- mean -> all queenright beehives anomaly score averaged

```python
anomalies.investigate_anomaly(
    investigated_features_dataframe=queenless_features_dataframe,
    baseline_features_dataframe=queenright_features_dataframe,
    baseline_aggregation='mean',
    observation_period='join',
    anomaly_scores_from_dataframe = anomaly_scores_from_dataframe,
    plotly=False,
);
```
